## keywords.csv

```python
import pandas as pd
import ast

# Load the CSV file
file_path = "keywords.csv"  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Specify the column containing the nested JSON-like data
column_name = ['keywords']  # Replace with your actual column name

for i in column_name:
    # Convert the string representation of the JSON-like data into Python objects
    df[i] = df[i].apply(ast.literal_eval)
    
    # Flatten the JSON-like column
    # Explode the list of dictionaries into separate rows
    df = df.explode(i, ignore_index=True)
    
    # Normalize the JSON objects and merge them with the original DataFrame
    normalized_data = pd.json_normalize(df[i])
    normalized_data.rename(columns={'id':'word_id', 'name' : 'word'}, inplace =True )
    df = pd.concat([df.drop(columns=[i]), normalized_data], axis=1)

# Save or display the resulting DataFrame
df.dropna(inplace = True)
df.to_csv("updated_keywords.csv", index=False)
print(df.head())
```

do the same for production_companies, production countries and spoken languages. some of the data is missing, we replace the missing data with [].

```python
from functools import partial

def extract_ids(missing, key, value):
    if pd.isna(value) or value == '' or len(value) == 0:
        return [missing]
    try:
        # Safely parse the string into a list of dictionaries and extract the ids
        parsed_value = ast.literal_eval(value)
        return [item[key] for item in parsed_value]
    except:
        # In case of error (invalid format), return [-1]
        return [missing]


f_companey = partial(extract_ids,-1, 'id')
movies_md['production_companies'] = movies_md['production_companies'].apply(f_companey)

f_country = partial(extract_ids,'US','iso_3166_1')
movies_md['production_countries'] = movies_md['production_countries'].apply(f_country)

f_lang = partial(extract_ids,'en','iso_639_1')
movies_md['spoken_languages'] = movies_md['spoken_languages'].apply(f_lang)
```

## Popularity-based filtering

We use [IMDB's Weighted Rating formula](https://en.wikipedia.org/wiki/IMDb#Rankings). <br>
$$ W = \frac{R \cdot v + C \cdot m}{v + m} $$
where: <br>
* w is the weighted raiting. <br>
* R is the mean rating for the movie, from 1 to 10.
* v is the number of votes for the movie.
* m is the minimum votes required to be listed (We chose 99% mark which is 2184)
* C is the mean vote across the whole report (110).

In our case, the lowest score possible is 110.

```python
pop_df = movies_md.loc[:,['id','vote_average','vote_count']]
pop_df.head()
```

```python
pop_df.isna().sum()
```

```python
pop_df.dropna(inplace = True)
pop_df.describe(),pop_df.vote_count.quantile(0.99)
```

```python
def w_rating(r , v ):
    c = 110
    m = 2184
    return ((r*v) + (c*m)) / (v+m)
```

```python
w_rating(0,0)
```

```python
pop_df['w_rating'] = w_rating(pop_df['vote_average'], pop_df['vote_count'])
```

for faster calculation we only use first 50000 movies

```python
#binary_matrix = k_merged.pivot_table(index='id', columns='word', aggfunc='size', fill_value=0)
binary_matrix = k_words[:100000].pivot_table(index='id', columns='word', aggfunc='size', fill_value=0)
binary_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity
cosine_sim = cosine_similarity(binary_matrix)

# Convert to DataFrame for better readability
cosine_sim_df = pd.DataFrame(cosine_sim, index=binary_matrix.index, columns=binary_matrix.index)
cosine_sim_df
```

```python
def recommend_movies(movie_id, df, top_n=3):
    # Get the similarity scores for the given movie
    sim_scores = df[movie_id].sort_values(ascending=False)

    # Exclude the input movie itself from the recommendations
    sim_scores = sim_scores[sim_scores.index != movie_id]

    # Get the top N most similar movies
    return sim_scores.head(top_n)

# Example usage: Recommend top 3 similar movies to movie 5

movie = 11
raw_md.loc[raw_md.id == str(movie)].title, raw_md.loc[raw_md.id.isin(recommend_movies(movie, cosine_sim_df, top_n=7).index.astype(str))].title
```

```python
rating_s
```

```python
user_movie_matrix = rating_s.pivot_table(index='userId', columns='movieId', values='rating', fill_value=0)
user_movie_matrix
```

### user based

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate similarity between users
user_similarity = cosine_similarity(user_movie_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)
user_similarity_df
```

```python
recommend_movies(1, user_similarity_df, top_n=10)
```

### movie based

```python
# Calculate similarity between movies
movie_similarity = cosine_similarity(user_movie_matrix.T)
movie_similarity_df = pd.DataFrame(movie_similarity, index=user_movie_matrix.columns, columns=user_movie_matrix.columns)
movie_similarity_df
```

```python
recommend_movies(1, movie_similarity_df, top_n=5)
```

```python
movie = 11
raw_md.loc[raw_md.id == str(movie)].title, raw_md.loc[raw_md.id.isin(recommend_movies(movie, movie_similarity_df, top_n=10).index.astype(str))].title
```

# SVD

```python
from sklearn.decomposition import TruncatedSVD

# Apply SVD
svd = TruncatedSVD(n_components=2)
decomposed_matrix = svd.fit_transform(user_movie_matrix.T)

# Compute similarity between users in the reduced space
movie_similarity_svd = cosine_similarity(decomposed_matrix)
movie_similarity_svd_df = pd.DataFrame(movie_similarity_svd, index=user_movie_matrix.columns, columns=user_movie_matrix.columns)
movie_similarity_svd_df
```

```python
movie = 11
raw_md.loc[raw_md.id == str(movie)].title, raw_md.loc[raw_md.id.isin(recommend_movies(movie, movie_similarity_svd_df, top_n=20).index.astype(str))].title
```

# hybrid

```python
def hybrid_recommend(movie_id, content_df, collab_df, alpha=0.5, top_n=5):
    
    # Get content-based similarity scores for the movie
    content_scores = content_df[movie_id]
    
    # Get collaborative similarity scores for the movie
    collab_scores = collab_df[movie_id]
    
    # Combine scores using a weighted average
    hybrid_scores = (alpha * content_scores) + ((1 - alpha) * collab_scores)
    
    # Sort movies by hybrid scores
    recommendations = hybrid_scores.sort_values(ascending=False).drop(movie_id)
    
    # Return top N recommendations
    return recommendations.head(top_n)

# Example: Recommend movies similar to movie 101
hybrid_recommend(101, cosine_sim_df, movie_similarity_df, alpha=0.5, top_n=10)

movie = 11
raw_md.loc[raw_md.id == str(movie)].title, raw_md.loc[raw_md.id.isin(hybrid_recommend(movie, cosine_sim_df, movie_similarity_df, alpha=0.5, top_n=10).index.astype(str))].title
```
